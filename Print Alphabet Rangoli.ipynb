### we take an input of number from user and using ASCII we make a beautiful pattern of lower alphabets




```python
def alpha_pattern(row_num, input_size):
    if row_num==0:
        print(chr(96+input_size),end='')
                
    else:
        for i in range(row_num+1):
            print(chr(96+input_size-i)+ '-',end='')
            right_start_bit = 96+input_size-i
        
        for i in range(row_num):
            print(chr(right_start_bit+i+1),end='')
            if i in range(row_num-1):
                print('-',end='')

        
def dash_pattern(size_cont_b,h_width,loop):
   
        print('-'*(h_width-size_cont_b),end='')
        if loop == 0:

            print(end='\n')


def print_rangoli(size):

    total_row = 2*size -1    #number of rows in rangoli
    half_width = int((total_row*2-2)/2)   # width of rangoli
    size_control_bit = 0
    lower_half_cont = 0
    
    for row in range(size):  # outer loop for each row 
        dash_pattern(size_control_bit,half_width,loop=1)
        alpha_pattern(row, size)
        dash_pattern(size_control_bit,half_width,loop=0)
        size_control_bit += 2
        lower_half_cont +=1
    
    size_control_bit -= 4
    for row_1 in range(size-1):  # outer loop for each row 
        
        dash_pattern(size_control_bit,half_width,loop=1)
        alpha_pattern(lower_half_cont-2, size)
        dash_pattern(size_control_bit,half_width,loop=0)
        size_control_bit -=2
        lower_half_cont -=1



if __name__ == '__main__':
    N = int(input())
    print_rangoli(N)
```
